```python
import matplotlib.pyplot  as plt
```

```python
x=[2,6]
y=[4,9]

x1=[8,3]
y1=[4,7]
plt.xlabel("time")
plt.ylabel("students")
plt.plot(x,y,label="ml learning")
plt.plot(x1,y1,label="cloud learning")
plt.legend()
plt.grid(color='g')
plt.show()
```

```python
x=[200,60,34,90]
y=["suku","priyanshi","big bro","lil bro"]
plt.xlabel("time")
plt.ylabel("speed")
plt.bar(y,x,label="coder",color='y')
plt.legend()
plt.grid(color='y')
plt.show()
```

```python
x=[2,6,5,6,7,78,9]
y=[4,9,23,6,9,6,5]

x1=[2,9,5,6,2,38,8]
y1=[5,8,24,5,8,2,5]

plt.xlabel("time")
plt.ylabel("speed")
plt.scatter(x,y,label="ml learning",marker='x',s=120,c='r')
plt.scatter(x1,y1,label="hadoop",s=150)
plt.legend()
plt.grid(color='g')
plt.show()
```
